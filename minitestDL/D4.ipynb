```python
dummies = pd.get_dummies(df['Location'])
df_onehot = pd.concat([df,dummies], axis = 1)
df_onehot.drop("Location",axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#for name in data.columns:
df_onehot['WindGustDir'] = le.fit_transform(df_onehot['WindGustDir'])
print(df_onehot['WindGustDir'])
```

```python
df_onehot['WindDir9am'] = df_onehot['WindDir9am'].astype(str)
df_onehot['WindDir9am'] = le.fit_transform(df_onehot['WindDir9am'])
```

```python
df_onehot.drop("Date",axis=1,inplace=True)
```

```python
df_onehot['RainToday'] = le.fit_transform(df_onehot['RainToday'])
```

```python
df_onehot['RainTomorrow'] = le.fit_transform(df_onehot['RainTomorrow'])
```

```python
df_onehot.info()
```

```python
from sklearn.model_selection import train_test_split
X = df_onehot.drop(['RainTomorrow'],axis=1)
Y = df_onehot['RainTomorrow']
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
#ANN
import tensorflow as tf
from tensorflow import keras
```

```python
X_train.shape
```

```python
model = keras.models.Sequential()
activation = 'relu'
model.add(keras.layers.Flatten(input_shape=[69]))
model.add(keras.layers.Dense(69,activation=activation))
model.add(keras.layers.Dense(35,activation=activation))
model.add(keras.layers.Dense(1,activation="sigmoid"))

model.summary()
```

```python
model.compile(loss="BinaryCrossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
X_train
```

```python
history = model.fit(X_train,y_train,validation_data = (X_test,y_test),epochs = 30)
```

```python
plt.plot(history.history["accuracy"],c='b')
plt.plot(history.history["val_accuracy"],c='r')
plt.grid()
plt.legend(["Training Acc","Validation Acc"])

import sklearn.metrics as metrics
y_pred = model.predict(X_test)
#print(y_pred)
y_pred = [1 if i > 0.5 else 0 for i in y_pred]

acc = metrics.accuracy_score(y_test,y_pred)
precision = metrics.precision_score(y_test,y_pred)
recall = metrics.recall_score(y_test,y_pred)

print(f'accuracy : {acc}')
print(f'precision : {precision}')
print(f'recall : {recall}')
```
